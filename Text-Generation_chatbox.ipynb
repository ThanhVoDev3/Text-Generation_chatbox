```python
from transformers import GPT2Tokenizer, TFGPT2LMHeadModel
```

```python
import gradio as gr
```

**01. Load Model**

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
model = TFGPT2LMHeadModel.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id)
```

**02. Tokenize Setences**

```python
sentence = "Vo Van Thanh I Live In Binh Phuoc"
input_ids = tokenizer.encode(sentence, return_tensors='pt')
```

```python
tokenizer.decode(input_ids[0])
```

```python
tokenizer.decode(input_ids[0][1])
```

**03. Generate and Decode Text**

```python
output = model.generate(input_ids,
                        max_length = 500,
                        num_beams = 5,
                        no_repeat_ngram_size=2,
                        early_stopping=True)
```

```python
output
```

```python
decoded_output = tokenizer.decode(output[0])
print(decoded_output)
```

```python
tokenizer.decode(output[0], skip_special_tokens=True)
```

**04.Output Result***

```python
import gradio as gr
```

```python
def generate_text(inp):
    # Limit the number of tokens in the input sentence
    max_input_length = 50
    inp = " ".join(inp.split()[:max_input_length])

    input_ids = tokenizer.encode(inp, return_tensors='tf')
    beam_output = model.generate(input_ids, max_length=100, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
    output = tokenizer.decode(beam_output[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
    return ".".join(output.split(".")[:-1]) + "."

output_text = gr.outputs.Textbox()
```

```python
input_text = gr.inputs.Textbox(lines=2, label="Nhập Câu hỏi:")

output_text = gr.outputs.Textbox(label="Câu trả lời:")

iface = gr.Interface(fn=generate_text, inputs=input_text, outputs=output_text)

iface.launch()
```
